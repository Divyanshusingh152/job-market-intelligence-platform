```python
import pandas as pd
import warnings
warnings.filterwarnings('ignore')

# Business question: What does our data look like?

df = pd.read_csv('Data/raw/marketing_sample_for_naukri_com-jobs__30k_data.csv')
print(f"Total job listings loaded: {len(df)}")
print(f"Total columns: {len(df.columns)}")
print(f"\nColumn names: {list(df.columns)}")
df.head(3)
```

```python
# Business question: Is our data healthy? Where are problems hiding?

print("=== MISSING VALUES (nulls) per column ===")
print(df.isnull().sum())

print("\n=== TOTAL ROWS:", len(df), "===")
print("Rows with at least one missing value:", df.isnull().any(axis=1).sum())

print("\n=== DATA TYPE of each column ===")
print(df.dtypes)
```

```python
list(df.columns)
```

```python
# Business question: How do we fix dirty data so we can trust our analysis?

# Step 1 — Rename columns (remove spaces, easier to type)
df.columns = ['id', 'crawl_date', 'job_title', 'salary',
              'experience', 'key_skills', 'role_category',
              'location', 'functional_area', 'industry', 'role']

# Step 2 — Drop rows where the 3 most important columns are missing
rows_before = len(df)
df.dropna(subset=['job_title', 'location', 'industry'], inplace=True)
rows_after = len(df)
print(f"Rows removed due to nulls: {rows_before - rows_after}")
print(f"Clean rows remaining: {rows_after}")

# Step 3 — Clean extra spaces and duplicates in location
# "Pune,Pune" → "Pune"
df['location'] = df['location'].str.strip()
df['location'] = df['location'].apply(
    lambda x: ', '.join(dict.fromkeys([c.strip() for c in x.split(',')])) if pd.notnull(x) else x
)

# Step 4 — Extract numbers from experience text
# "5 - 10 yrs" → min_exp = 5, max_exp = 10
df['exp_min'] = df['experience'].str.extract(r'(\d+)').astype(float)
df['exp_max'] = df['experience'].str.extract(r'\d+\s*-\s*(\d+)').astype(float)

# Step 5 — Fix the date column (text → proper date)
df['crawl_date'] = pd.to_datetime(df['crawl_date'], utc=True)

# Step 6 — Confirm everything looks right
print("\n=== NULLS AFTER CLEANING ===")
print(df[['job_title','location','industry','exp_min','exp_max']].isnull().sum())

print("\n=== SAMPLE OF EXPERIENCE EXTRACTION ===")
print(df[['experience','exp_min','exp_max']].dropna().head(5))

print("\n=== DATE COLUMN FIXED ===")
print(df['crawl_date'].dtype)
print(df['crawl_date'].head(3))
```

```python
# Business question: Where are the jobs? 
# Which cities and industries are hiring most?
# Business question: Where are the jobs? 
# Which cities and industries are hiring most?

print("=== TOP 10 CITIES HIRING ===")
print(df['location'].value_counts().head(10))

print("\n=== TOP 10 INDUSTRIES ===")
print(df['industry'].value_counts().head(10))

print("\n=== TOP 10 JOB TITLES ===")
print(df['job_title'].value_counts().head(10))

print("\n=== EXPERIENCE LEVEL SUMMARY ===")
print(f"Average minimum experience required: {df['exp_min'].mean():.1f} years")
print(f"Most common experience range: {df['experience'].value_counts().index[0]}")
```

```python
# Business question: What are the most in-demand skills 
# across all 30,000 job listings RIGHT NOW?

# Step 1 — Drop rows where skills are missing
skills_df = df['key_skills'].dropna()

# Step 2 — Split pipe-separated skills into individual items
# "Python| SQL| Excel" → ['Python', 'SQL', 'Excel']
skills_series = skills_df.str.split('|')

# Step 3 — explode() turns each skill into its own row
# One row with 3 skills becomes 3 rows with 1 skill each
skills_exploded = skills_series.explode()

# Step 4 — Clean spaces and fix capitalisation
skills_exploded = skills_exploded.str.strip().str.title()

# Step 5 — Count every skill
top_skills = skills_exploded.value_counts().head(20)

print("=== TOP 20 IN-DEMAND SKILLS (from 30,000 job listings) ===\n")
for rank, (skill, count) in enumerate(top_skills.items(), 1):
    bar = '█' * (count // 200)
    print(f"{rank:2}. {skill:<35} {count:>5}  {bar}")
```

```python
# Business question: Is this market fresher-friendly or 
# does everyone want experienced candidates?

# Step 1 — Categorise jobs by experience level
bins = [-1, 1, 3, 6, 10, 50]
labels = ['Fresher (0-1 yr)', 'Junior (1-3 yrs)', 
          'Mid (3-6 yrs)', 'Senior (6-10 yrs)', 
          'Expert (10+ yrs)']

df['exp_level'] = pd.cut(df['exp_min'], bins=bins, labels=labels)

# Step 2 — Count jobs per experience level
exp_counts = df['exp_level'].value_counts().sort_index()

print("=== JOB MARKET BY EXPERIENCE LEVEL ===\n")
total = exp_counts.sum()
for level, count in exp_counts.items():
    pct = (count / total) * 100
    bar = '█' * int(pct // 2)
    print(f"{level:<22} {count:>5} jobs  ({pct:.1f}%)  {bar}")

# Step 3 — Fresher friendly cities
print("\n=== TOP 5 FRESHER-FRIENDLY CITIES ===")
fresher_df = df[df['exp_min'] <= 1]
print(fresher_df['location'].value_counts().head(5))

# Step 4 — Fresher friendly industries  
print("\n=== TOP 5 FRESHER-FRIENDLY INDUSTRIES ===")
print(fresher_df['industry'].value_counts().head(5))
```

```python
# Push clean dataframe into PostgreSQL as a table
df.to_sql(
    name='job_listings',      # table name in database
    con=engine,               # connection we created
    if_exists='replace',      # if table exists, replace it
    index=False,              # don't write row numbers as a column
    dtype={                   # tell postgres what type each column is
        'id': sqlalchemy.Text(),
        'crawl_date': sqlalchemy.DateTime(timezone=True),
        'job_title': sqlalchemy.Text(),
        'salary': sqlalchemy.Text(),
        'experience': sqlalchemy.Text(),
        'key_skills': sqlalchemy.Text(),
        'role_category': sqlalchemy.Text(),
        'location': sqlalchemy.Text(),
        'functional_area': sqlalchemy.Text(),
        'industry': sqlalchemy.Text(),
        'role': sqlalchemy.Text(),
        'exp_min': sqlalchemy.Float(),
        'exp_max': sqlalchemy.Float(),
        'exp_level': sqlalchemy.Text(),
    }
)

print("Table created successfully!")
print(f"Total rows pushed: {len(df)}")
```

```python
# SQL Query 1
# Business question: Which city has the most job opportunities?

query1 = """
SELECT 
    location,
    COUNT(*) as job_count
FROM job_listings
GROUP BY location
ORDER BY job_count DESC
LIMIT 10;
"""

result1 = pd.read_sql(query1, engine)
print("=== TOP 10 CITIES BY JOB COUNT ===\n")
print(result1.to_string(index=False))
```

```python
# SQL Query 2
# Business question: Which industry hires most freshers?

query2 = """
SELECT 
    industry,
    COUNT(*) as fresher_jobs
FROM job_listings
WHERE exp_min <= 1
GROUP BY industry
ORDER BY fresher_jobs DESC
LIMIT 10;
"""

result2 = pd.read_sql(query2, engine)
print("=== TOP 10 FRESHER FRIENDLY INDUSTRIES ===\n")
print(result2.to_string(index=False))
```

```python
# SQL Query 3
# Business question: Which industries are hardest to enter?

query3 = """
SELECT 
    industry,
    ROUND(AVG(exp_min)::numeric, 1) as avg_experience,
    COUNT(*) as total_jobs
FROM job_listings
WHERE exp_min IS NOT NULL
GROUP BY industry
HAVING COUNT(*) > 100
ORDER BY avg_experience DESC
LIMIT 10;
"""

result3 = pd.read_sql(query3, engine)
print("=== INDUSTRIES REQUIRING HIGHEST EXPERIENCE ===\n")
print(result3.to_string(index=False))
```

```python
# SQL Query 4
# Business question: Which city + industry combination 
# has the most job opportunities?

query4 = """
SELECT 
    location,
    industry,
    COUNT(*) as job_count
FROM job_listings
GROUP BY location, industry
ORDER BY job_count DESC
LIMIT 10;
"""

result4 = pd.read_sql(query4, engine)
print("=== TOP 10 CITY + INDUSTRY COMBINATIONS ===\n")
print(result4.to_string(index=False))
```

```python
# SQL Query 5
# Business question: Which industries have more than 500 job listings?

query5 = """
SELECT 
    industry,
    COUNT(*) as job_count
FROM job_listings
GROUP BY industry
HAVING COUNT(*) > 500
ORDER BY job_count DESC;
"""

result5 = pd.read_sql(query5, engine)
print("=== INDUSTRIES WITH MORE THAN 500 JOB LISTINGS ===\n")
print(result5.to_string(index=False))
```

```python
# SQL Query 6
# Business question: Which job roles have national demand
# appearing in more than 5 different cities?

query6 = """
SELECT 
    role,
    COUNT(DISTINCT location) as city_count,
    COUNT(*) as total_jobs
FROM job_listings
WHERE role IS NOT NULL
GROUP BY role
HAVING COUNT(DISTINCT location) > 5
ORDER BY city_count DESC
LIMIT 10;
"""

result6 = pd.read_sql(query6, engine)
print("=== TOP 10 ROLES WITH NATIONAL DEMAND ===\n")
print(result6.to_string(index=False))
```

```python
# SQL Query 7
# Business question: Rank all industries by job count
# Which is #1, #2, #3?

query7 = """
SELECT 
    industry,
    COUNT(*) as job_count,
    RANK() OVER (ORDER BY COUNT(*) DESC) as industry_rank
FROM job_listings
GROUP BY industry
ORDER BY industry_rank
LIMIT 10;
"""

result7 = pd.read_sql(query7, engine)
print("=== INDUSTRIES RANKED BY JOB COUNT ===\n")
print(result7.to_string(index=False))
```

```python
# SQL Query 8
# Business question: Which cities have highest 
# percentage of fresher jobs?

query8 = """
WITH city_total AS (
    SELECT 
        location, 
        COUNT(*) as total
    FROM job_listings
    GROUP BY location
),
city_fresher AS (
    SELECT 
        location, 
        COUNT(*) as fresher_jobs
    FROM job_listings
    WHERE exp_min <= 1
    GROUP BY location
)
SELECT 
    t.location,
    t.total,
    f.fresher_jobs,
    ROUND(f.fresher_jobs::numeric / t.total * 100, 1) as fresher_pct
FROM city_total t
JOIN city_fresher f ON t.location = f.location
WHERE t.total > 200
ORDER BY fresher_pct DESC
LIMIT 10;
"""

result8 = pd.read_sql(query8, engine)
print("=== CITIES WITH HIGHEST FRESHER JOB PERCENTAGE ===\n")
print(result8.to_string(index=False))
```
